Exercise 1: Function Plotting and Analytical Determination of Critical Points

```python

# Import numpy and pyplot libraries
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f1(x, y):
    return x**3 + y**3 - 3*x - 12*y

def f2(x, y):
    return x**3 - 12*x*y + 8*y**3

def f3(x, y):
    return (x - 1)**2 + 10*(x**2 - y)**2
```

1. Fonction $f_1(x,y) = x^3 + y^3 - 3x - 12y$

A. Points Critiques ($\nabla f_1 = 0$)Gradient (Dérivées Premières) :$$\frac{\partial f_1}{\partial x} = 3x^2 - 3$$$$\frac{\partial f_1}{\partial y} = 3y^2 - 12$$Résolution :

$3x^2 - 3 = 0 \implies x = \pm 1$$
3y^2 - 12 = 0 \implies y = \pm 2$

Coordonnées des Points Critiques :$$P_1 = (1, 2), \quad P_2 = (1, -2), \quad P_3 = (-1, 2), \quad P_4 = (-1, -2)$$


2. Fonction $f_2(x,y) = x^3 - 12xy + 8y^3$

A. Points Critiques ($\nabla f_2 = 0$)

Gradient (Dérivées Premières) :$$\frac{\partial f_2}{\partial x} = 3x^2 - 12y$$$$\frac{\partial f_2}{\partial y} = -12x + 24y^2$$

Résolution du Système :

$y = \frac{x^2}{4}$

$x = 2y^2$

Substitution $\implies y^4 - y = 0 \implies y(y^3 - 1) = 0$.

Coordonnées des Points Critiques :$$P_1 = (0, 0), \quad P_2 = (2, 1)$$

3. Fonction $f_3(x,y) = (x-1)^2 + 10(x^2 - y)^2$

A. Points Critiques ($\nabla f_3 = 0$)

Gradient (Dérivées Premières) :$$\frac{\partial f_3}{\partial x} = 2(x-1) + 40x(x^2 - y)$$$$\frac{\partial f_3}{\partial y} = -20(x^2 - y)$$

Résolution du Système :

L'équation la plus simple est :$$\frac{\partial f_3}{\partial y} = 0 \implies -20(x^2 - y) = 0 \implies y = x^2$$

On substitue $y = x^2$ dans la première équation :$$\frac{\partial f_3}{\partial x} = 0 \implies 2(x-1) + 40x(\underbrace{x^2 - x^2}_{0}) = 0$$$$2(x-1) = 0 \implies x = 1$$

Coordonnées du Point Critique :

Si $x=1$, alors $y = 1^2 = 1$. $\implies P_c = (1, 1)$.

```python
# Definition of the plotting domain
xmin, xmax, nx = -4, 4, 41
ymin, ymax, ny = -4, 4, 41
# Discretization of the plotting domain
x1d = np.linspace(xmin,xmax,nx)
y1d = np.linspace(ymin,ymax,ny)
x2d, y2d = np.meshgrid(x1d, y1d)

# Plot contour lines (isovalues) of f1
nIso = 21
plt.contour(x2d,y2d,f1(x2d,y2d),nIso)
plt.title('Contour Lines')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.axis('square')
```

```python
# Intervalle 2 : [-2, 2] (zoom local + détails)
xmin, xmax, nx = -2, 2, 101   # On met plus de points pour un meilleur zoom
ymin, ymax, ny = -2, 2, 101

x1d = np.linspace(xmin, xmax, nx)
y1d = np.linspace(ymin, ymax, ny)
x2d, y2d = np.meshgrid(x1d, y1d)

nIso = 21
plt.contour(x2d, y2d, f1(x2d, y2d), nIso)
plt.title('Contour Plot of f1 on [-2,2] × [-2,2] (zoom)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.axis('square')
plt.show()
```

###What happens if the domain discretization is insufficient?
If the discretization is too low, the function is not sampled enough.
The contour lines become imprecise or wrong, and important details (like minima or saddle points) may disappear.
In short: low discretization = inaccurate and misleading plot
